# Part 1

model growth rate of lanternfish population
- each lanternfish creates a new one once every 7 days
- you can model each fish as a single number that represents the number of days until it creates a new fish
- first cycle: 2 more days (so 9 days)
- start at 8, go down to 0, then reset to 6 (but add a new one)


```python
def simulate_days(days, ages):
    for day in range(1, days+1):
        new_fish = ages.count(0)
        ages = [a-1 if a>0 else 6 for a in ages]
        ages += [8] * new_fish
    print(f"After {day} days: {len(ages)}")

def part1(file_name, days):
    with open(file_name, 'r') as f:
        lines = f.readlines()
        ages = [int(entry) for entry in lines[0].strip().split(',')]
    print(ages)
    simulate_days(days, ages)
```

```python
part1('example.txt', 80)
```

```python
part1('input.txt', 80)
```

Next idea, don't save every fish in the list separately, instead save the number of fish in each stage

[day0, day1, day2, day3, day4, day5, day6, day7, day8]

then each day: 
- take the day0s
- shift everything to the left once
- add the day0 to day6 and day8

```python
def simulate_days2(days, fish_at_stage):
    for day in range(1, days+1):
        expired_fish = fish_at_stage.pop(0)
        fish_at_stage[6] += expired_fish
        fish_at_stage.append(expired_fish)
        #print(f'After day {day}: {fish_at_stage}')
    print(f"After {day} days: {sum(fish_at_stage)}")

def part2(file_name, days):
    with open(file_name, 'r') as f:
        lines = f.readlines()
        ages = [int(entry) for entry in lines[0].strip().split(',')]
        fish_at_stage = [ages.count(i) for i in range(9)]
    print('ages/input', ages)
    print('fish at stage', fish_at_stage)
    simulate_days2(days, fish_at_stage)
```
